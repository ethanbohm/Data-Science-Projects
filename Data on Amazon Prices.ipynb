```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f_oneway
from statsmodels.stats.multicomp import pairwise_tukeyhsd
from IPython.display import display
```

```python
amazon = pd.read_csv("amazon.csv")

# replace str values, convert to numeric, and convert to US Dollars for both actual and discounted

amazon["actual_price"] = amazon.actual_price.str.replace("₹", "")
amazon["actual_price"] = amazon.actual_price.str.replace(",", "")
amazon["actual_price"] = pd.to_numeric(amazon.actual_price)
amazon["actual_price"] = amazon.actual_price * 0.012

amazon["discounted_price"] = amazon.discounted_price.str.replace("₹", "")
amazon["discounted_price"] = amazon.discounted_price.str.replace(",", "")
amazon["discounted_price"] = pd.to_numeric(amazon.discounted_price)
amazon["discounted_price"] = amazon.discounted_price * 0.012

# split categories by first two names in str (Makes data more useable)

amazon["official_category"] = amazon.category.str.split("|").str[0]

# print number of new unique categories

display(amazon.head())
print(amazon.official_category.unique())

# compare average actual prices of each category

grouped = amazon.groupby('official_category')['actual_price'].mean()

plt.bar(grouped.index, grouped.values)
plt.xticks(rotation=90)
plt.title("Average Price per Category")
plt.xlabel('Category')
plt.ylabel('Average Price')
plt.show()

# Now lets run an ANOVA A/B test to find out if there are any significant differences in price

# perform one-way ANOVA
f_stat, p_value = f_oneway(amazon[amazon.official_category == 'Computers&Accessories']['actual_price'],
                                 amazon[amazon.official_category == 'Electronics']['actual_price'],
                                 amazon[amazon.official_category == 'Home&Kitchen']['actual_price'])

# print the p-value
print("ANOVA p-value:", p_value)

# perform Tukey HSD test
tukey_results = pairwise_tukeyhsd(amazon.actual_price, amazon.official_category)

# print the results
print(tukey_results)

print("* We find that there is a significant difference in price between \n Electronics and Computers&Accessories, Home&Kitchen, and Office Products.")

```
